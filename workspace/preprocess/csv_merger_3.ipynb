```python
import pandas as pd

df = pd.read_csv('/home/chio4696/DLthon_3team/DATA/train/non_casual/gpt_augmented_non_casual.csv')

df = df.drop(columns=['conversation'])
```

```python
df
```

```python
class1 = []
conversation = []

for idx, row in df.iterrows():
    class1.append(row['class'])
    conversation.append(row['cynical'])
    
    class1.append(row['class'])
    conversation.append(row['informal'])
    
    class1.append(row['class'])
    conversation.append(row['formal'])
    
    class1.append(row['class'])
    conversation.append(row['concise'])
```

```python
new_df = pd.DataFrame({
    'class': class1,
    'conversation': conversation
})

new_df
```

```python
new_df['idx'] = range(0, len(new_df))

new_df = new_df[["idx", "class", "conversation"]]

new_df.to_csv('/home/chio4696/DLthon_3team/DATA/train/non_casual/gpt_augmented_non_casual2.csv', index=False)
```
